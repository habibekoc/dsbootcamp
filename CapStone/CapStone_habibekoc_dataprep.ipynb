```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from dateutil import parser
from google.colab import files
```

```python
df_t = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/transactions.csv')
```

```python
df_c = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/customers.csv')
df_c
```

```python
df_t['t_dat'] = pd.to_datetime(df_t['t_dat'], format='%Y-%m-%d')
```

```python
max_t = df_t['t_dat'].max()
last_month = max_t - timedelta(days=31)
df_lm = df_t[df_t['t_dat'] > (last_month)]
```

```python
del df_lm['price']
del df_lm['sales_channel_id']
del df_lm['t_dat']
```

```python
df_lm
```

```python
df_count = df_lm.groupby(['customer_id', 'article_id']).size().to_frame(name = 'count').reset_index()
df_count
```

```python
temp = df_c[['customer_id','age']]
temp
```

```python
df_count = pd.merge(df_count, temp, how='inner')
df_count
```

```python
bins = [15, 25, 35, 45, 55, 65, 75, 85, 100]
labels = [1, 2, 3, 4, 5, 6, 7, 8]
df_count['binned_age'] = pd.cut(df_count['age'], bins=bins, labels=labels)
```

```python
del df_count['age']
```

```python
df_count
```

```python
df_count.to_csv('small_transactions1.csv', encoding = 'utf-8-sig') 
files.download('small_transactions1.csv')
```
